```python
from pathlib import Path

# Change this to your folder path
folder = Path(r"C:/Users/Anwender/Documents/Epicode_Python/co-occurence_prepararsi_al_presente/works/works_no_consequences_aftermaths")

# Find all .yaml and .yml files
files = list(folder.glob("*.yaml")) + list(folder.glob("*.yml"))

for file in files:
    with open(file, "r", encoding="utf-8") as f:
        lines = f.readlines()

    # Remove line 10, 11, 12 (Python index 9, 10, 11)
    if len(lines) >= 11:
        del lines[9:11]

        with open(file, "w", encoding="utf-8") as f:
            f.writelines(lines)

        print(f"Updated: {file.name}")
    else:
        print(f"Skipped (less than 12 lines): {file.name}")
```
